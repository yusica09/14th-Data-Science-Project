### Random Forests는 Decision Tree 모델을 여러개 만든 것

### 새로 들어온 고객이 물건을 구매할지, 구매하지 않을지를 분류하는 문제

```python
from sklearn import tree
X = [[0, 0], [1, 1]]
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
# fit을 통해 학습
clf = clf.fit(X, Y)
```

```python
clf.predict([[2., 2.]])
```

```python
# predict_proba => 비율로 출력
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
X, y
```

* 위의 0,1,2는 카테고리 분류용

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
tree.plot_tree(clf.fit(X,y), filled=True)
```

* decision tree = 의사 결정 나무 (특정 조건에 따른 트리)
* Decision tree는 CART(Classification and Regression Tree) 알고리즘이라고도 불림
